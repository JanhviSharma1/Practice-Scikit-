```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample Data
x = np.arange(1, 6)  # X values [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]  # Y values

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)  # 2 Rows, 3 Columns, Position 1
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Plot")
plt.legend()
plt.grid(True)

# 2. Scatter Plot
plt.subplot(2, 3, 2)  # Position 2
plt.scatter(x, y, color='r', label="Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")
plt.legend()
plt.grid(True)

# 3. Bar Chart
plt.subplot(2, 3, 3)  # Position 3
plt.bar(x, y, color='g', label="Bar Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bar Chart")
plt.legend()
plt.grid(axis='y')

# 4. Histogram
data = np.random.randn(1000)  # Random data
plt.subplot(2, 3, 4)  # Position 4
plt.hist(data, bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")

# 5. Pie Chart
labels = ['A', 'B', 'C', 'D', 'E']
plt.subplot(2, 3, 5)  # Position 5
plt.pie(y, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'yellow', 'orange'])
plt.title("Pie Chart")

# Show All Plots
plt.tight_layout()
plt.show()
```

```python
# Sample Data
x = np.linspace(1, 10, 10)  # Values from 1 to 10
y1 = x ** 2  # Quadratic function
y2 = np.sqrt(x)  # Square root function
data = np.random.randn(1000)  # Random data for histogram & boxplot

# Create a Figure with Multiple Subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 8))  # 2 rows, 3 columns

### 1. Multiple Line Plot
axes[0, 0].plot(x, y1, marker='o', linestyle='-', color='b', label="y = x^2")
axes[0, 0].plot(x, y2, marker='s', linestyle='--', color='r', label="y = x")
axes[0, 0].set_xlabel("X-Axis")
axes[0, 0].set_ylabel("Y-Axis")
axes[0, 0].set_title("Multiple Line Plot")
axes[0, 0].legend()
axes[0, 0].grid(True)

### 2. Scatter Plot with Color Mapping
colors = np.random.rand(10)  # Random colors for each point
axes[0, 1].scatter(x, y1, c=colors, cmap='viridis', s=100, edgecolors='black')
axes[0, 1].set_title("Scatter Plot with Colormap")
axes[0, 1].set_xlabel("X")
axes[0, 1].set_ylabel("Y")

### 3. Histogram (Data Distribution)
axes[0, 2].hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')
axes[0, 2].set_title("Histogram of Random Data")
axes[0, 2].set_xlabel("Value")
axes[0, 2].set_ylabel("Frequency")

### 4. Box Plot (Statistical Summary)
axes[1, 0].boxplot(data, patch_artist=True, boxprops=dict(facecolor="lightblue"))
axes[1, 0].set_title("Box Plot of Data")

### 5. Heatmap (Using imshow)
heatmap_data = np.random.rand(10, 10)  # 10x10 matrix
im = axes[1, 1].imshow(heatmap_data, cmap='coolwarm', interpolation='nearest')
plt.colorbar(im, ax=axes[1, 1])  # Add colorbar
axes[1, 1].set_title("Heatmap")

### 6. Pie Chart (Data Distribution)
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 35, 15]
axes[1, 2].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
axes[1, 2].set_title("Pie Chart Representation")

# Adjust Layout and Save Figure
plt.tight_layout()
plt.savefig("data_analytics_visualization.png", dpi=300)  # Save as an image
plt.show()
```
