```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1, 2, 3, 4, 5, 6, 7])
y=np.array([2, 4, 6, 8, 10, 12, 14])
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

#another way
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()                           # Show the figure.
```

```python
x = [0, 2, 4, 6, 8]
y = [0, 4, 16, 36, 64]
fig, ax = plt.subplots()
ax.plot(x, y, marker='o', label="Data Points")
ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.legend(loc="best")
plt.show()
```

```python
# Sample data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]

fig, ax = plt.subplots(3, figsize=(8, 10))  # Create three subplots

ax[0].bar(labels, sizes)  # Vertical bar chart
ax[0].set_title("Bar Chart")

ax[1].barh(labels, sizes)  # Horizontal bar chart
ax[1].set_title("Horizontal Bar Chart")

ax[2].hist(sizes, bins=4, edgecolor='black')  # Histogram of sizes (keyword arguments to appear after given inputs)
ax[2].set_title("Histogram of Sizes")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
x=[1, 2, 3, 4]
y=[10, 20, 30, 40]
fig, ax = plt.subplots(2)  # Create two subplots
ax[0].plot(x, y)   # line chart
ax[1].scatter(x, y)  # scatter chart
plt.show()
```
